```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, MaxPooling1D, Flatten
from tensorflow.keras.optimizers import Adam
```

```python
df = pd.read_csv("^NSEI.csv")
df
```

```python
df = df.dropna()
df
#df = df.drop(['Adj Close'], axis = 1)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.set_index('Date', inplace=True)
```

```python
df = df.astype(float)
```

```python
df_seq = df.asfreq('B')
```

```python
train_data = df_seq.loc[:'2018-12-28']
test_data = df_seq.loc['2018-12-31':'2020-07-31']
test_data
```

```python
scaler = MinMaxScaler()
train_data_scaled = scaler.fit_transform(train_data)
test_data_scaled = scaler.transform(test_data)
```

```python
X_train = []
y_train = []
for i in range(5, len(train_data_scaled)):
    X_train.append(train_data_scaled[i-5:i, 0])
    y_train.append(train_data_scaled[i, 0])
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
model = Sequential()
model.add(Conv1D(filters=16, kernel_size=3, activation='relu', input_shape=(5, 1)))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(5, activation='relu'))
model.add(Dense(1, activation='linear'))
```

```python
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mse')
```

```python
epochs = 20
batch_size = 4
for i in range(epochs):
    for j in range(0, len(X_train) - batch_size, batch_size):
        x_train = X_train[j:j+batch_size, :]
        y_train_batch = y_train[j:j+batch_size]
        model.train_on_batch(x_train, y_train_batch)
```

```python
X_test = []
y_test = []
for i in range(5, len(test_data_scaled)):
    X_test.append(test_data_scaled[i-5:i, 0])
    y_test.append(test_data_scaled[i, 0])
X_test, y_test = np.array(X_test), np.array(y_test)
```

```python
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
predictions = model.predict(X_test)
predictions = scaler.inverse_transform(predictions)
```

```python
rmse = np.sqrt(np.mean((predictions - test_data[5:, 0])**2))
print('Mean RMSE:', rmse)
```
